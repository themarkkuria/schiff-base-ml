```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem, Draw, Descriptors
from IPython.display import display
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Load data from CSV
df = pd.read_csv("../data/molecules.csv")

# Display the loaded data for verification
display(df)

# Filter aldehydes and amines
aldehydes = df[df['type'] == 'aldehyde']
amines = df[df['type'] == 'amine']

# Visualize starting reactants
ald_mols = [Chem.MolFromSmiles(smi) for smi in aldehydes['smiles']]
amine_mols = [Chem.MolFromSmiles(smi) for smi in amines['smiles']]
reactant_mols = ald_mols + amine_mols
reactant_legends = list(aldehydes['name']) + list(amines['name'])

img_reactants = Draw.MolsToGridImage(
    reactant_mols,
    legends=reactant_legends,
    molsPerRow=4,
    subImgSize=(250, 250),
    useSVG=True
)
display(img_reactants)  # Show starting reactants

# Define the SMARTS for imine formation (Schiff base): aldehyde + primary amine -> imine
imine_rxn = AllChem.ReactionFromSmarts('[C:1]=[O:2].[NH2:3]>>[C:1]=[N:3]')

# Generate products automatically
auto_products = []
for _, ald_row in aldehydes.iterrows():
    ald_mol = Chem.MolFromSmiles(ald_row['smiles'])
    if ald_mol is None:
        print(f"Invalid aldehyde SMILES: {ald_row['smiles']}")
        continue
    for _, am_row in amines.iterrows():
        am_mol = Chem.MolFromSmiles(am_row['smiles'])
        if am_mol is None:
            print(f"Invalid amine SMILES: {am_row['smiles']}")
            continue
        ps = imine_rxn.RunReactants((ald_mol, am_mol))
        if ps:
            prod_mol = ps[0][0]  # Take the first product
            try:
                Chem.SanitizeMol(prod_mol)
                smi = Chem.MolToSmiles(prod_mol)
                auto_products.append({
                    'aldehyde': ald_row['name'],
                    'amine': am_row['name'],
                    'product': smi
                })
                print(f"Success: {ald_row['name']} + {am_row['name']} -> {smi}")
            except Exception as e:
                print(f"Sanitization error for {ald_row['name']} + {am_row['name']}: {e}")
        else:
            print(f"No product from {ald_row['name']} + {am_row['name']}")

# Create DataFrame from auto products
auto_df = pd.DataFrame(auto_products)
display(auto_df)

# Compute descriptors for generated Schiff bases
def compute_descriptors(smi):
    mol = Chem.MolFromSmiles(smi)
    if mol:
        Chem.SanitizeMol(mol)
        return {
            'MolWt': Descriptors.MolWt(mol),
            'logP': Descriptors.MolLogP(mol),
            'TPSA': Descriptors.TPSA(mol)
        }
    return {'MolWt': None, 'logP': None, 'TPSA': None}

auto_df = auto_df.join(auto_df['product'].apply(compute_descriptors).apply(pd.Series))

# Visualize each generated Schiff base individually with properties
if 'product' in auto_df.columns and not auto_df.empty:
    # Prepare features for ML (drop rows with None values from descriptor computation)
    ml_df = auto_df.dropna(subset=['MolWt', 'logP', 'TPSA'])

    # Toy training data (replace with real dataset, e.g., from ChEMBL)
    toy_data = pd.DataFrame({
        'MolWt': [200, 250, 300, 220, 280, 310, 240, 290, 260, 270],
        'logP': [2.5, 3.0, 1.8, 2.2, 3.5, 1.5, 2.8, 3.2, 2.0, 3.1],
        'TPSA': [50, 60, 40, 55, 65, 45, 52, 62, 48, 58],
        'solubility': [0.1, 0.05, 0.15, 0.08, 0.12, 0.04, 0.13, 0.06, 0.11, 0.07]  # mg/L as example
    })
    X_train, X_test, y_train, y_test = train_test_split(
        toy_data[['MolWt', 'logP', 'TPSA']], toy_data['solubility'], test_size=0.2, random_state=42
    )

    # Train Random Forest Regressor
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    # Evaluate model (optional)
    y_pred_test = model.predict(X_test)
    print(f"Model RÂ² Score on Test Set: {r2_score(y_test, y_pred_test):.2f}")

    # Predict solubility for generated Schiff bases using DataFrame with feature names
    X_predict = ml_df[['MolWt', 'logP', 'TPSA']]
    predictions = model.predict(X_predict)
    ml_df['predicted_solubility'] = predictions

    # Rank by predicted solubility
    ranked_df = ml_df.sort_values(by='predicted_solubility', ascending=False)

    # Display each ranked Schiff base with image and properties
    for _, row in ranked_df.iterrows():
        smi = row['product']
        mol = Chem.MolFromSmiles(smi)
        if mol:
            Chem.SanitizeMol(mol)
            Chem.Kekulize(mol, clearAromaticFlags=True)
            
            # Draw and display the individual molecule image
            legend = f"{row['aldehyde']} + {row['amine']} (Solubility: {row['predicted_solubility']:.2f} mg/L)"
            img = Draw.MolToImage(
                mol,
                size=(350, 350),
                legend=legend
            )
            display(img)
            
            # Display properties in a table
            prop_df = pd.DataFrame({
                'Property': ['Molecular Weight', 'logP', 'TPSA', 'Predicted Solubility'],
                'Value': [f"{row['MolWt']:.2f}", f"{row['logP']:.2f}", f"{row['TPSA']:.2f}", f"{row['predicted_solubility']:.2f}"]
            })
            display(prop_df)
        else:
            print(f"Skipping invalid molecule: {smi}")
else:
    print("No products generated. Check reaction inputs.")
```
